```python
import thinfilm as tm
import material as ml
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# access material paramater
SiO2_file = pd.read_csv('./SiO2.csv')
SiO2_w = SiO2_file['nm'].values
SiO2_n = SiO2_file['n'].values
SiO2_k = SiO2_file['k'].values
# material imported
SiO2 = ml.Material(SiO2_w, SiO2_n, SiO2_k, SiO2_w)
```

```python
ml.save_material(SiO2, 'SiO2') # material saved
```

```python
SiO2_saved = ml.open_material('SiO2', 'Material') # accese materail from saved
wl = np.linspace(400,700,301) # wavelength range

# figure plot
plt.plot(wl, SiO2_saved.nvalues(wl),'-k')
plt.xlabel('wavelength (nm)', fontsize = 14)
plt.ylabel('refractive index', fontsize = 14)
# plt.savefig('SiO2_n.png')
plt.show()
plt.plot(wl, SiO2_saved.kvalues(wl),'--b')
plt.xlabel('wavelength (nm)', fontsize = 14)
plt.ylabel('extinction coefficient', fontsize = 14)
# plt.savefig('SiO2_k.png')
plt.show()
```
